```python
# import des modules
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
# création du circuit et affichage de base
qCircuit = QuantumCircuit(8, 1)
qCircuit.draw('mpl', style="iqp")
```

```python
# Ajout d'une porte Hadamard sur chaque qbit pour représenter toutes les personnes
qCircuit.h(0)
qCircuit.h(1)
qCircuit.h(2)
qCircuit.h(3)
qCircuit.barrier()

qCircuit.draw('mpl', style="iqp")
```

```python
# Ajout des portes CNOT pour calculer le nombre pair ou impair de cheveux indigo
qCircuit.cx(1, 4)
qCircuit.cx(2, 4)
qCircuit.cx(3, 4)
qCircuit.barrier()

qCircuit.draw('mpl', style="iqp")
```

```python
# Représentation des 3 autres personnes qui entendent la réponse précédente
qCircuit.cx(4, 5)
qCircuit.cx(4, 6)
qCircuit.cx(4, 7)
qCircuit.barrier()

qCircuit.draw('mpl', style="iqp")
```

```python
# Bob voit les cheveux des personnes devant lui et déduis la couleur des siens
qCircuit.cx(2, 5)
qCircuit.cx(3, 5)
qCircuit.barrier()

qCircuit.draw('mpl', style="iqp")
```

```python
# Charlie et Dahlia prennent en compte le résultat de Bob
qCircuit.cx(5, 6)
qCircuit.cx(5, 7)
qCircuit.barrier()

qCircuit.draw('mpl', style="iqp")
```

```python
# Charlie voit la couleur de la chevelure devant lui et annonce sa propre couleur
qCircuit.cx(3, 6)
qCircuit.barrier()

qCircuit.draw('mpl', style="iqp")
```

```python
# Dahlia note cette réponse et annonce la couleur de sa chevelure
qCircuit.cx(6, 7)

qCircuit.draw('mpl', style="iqp")
```

```python
# Mesure des qBits
qCircuit.measure_all()
qCircuit.draw('mpl', style="iqp", fold=-1)
```

```python
# Simulation 
simulation = Aer.get_backend('qasm_simulator')
data = execute(qCircuit, backend=simulation, shots=10000).result()
plot_histogram(data.get_counts())
```
